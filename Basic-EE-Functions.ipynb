# A Few Basic Earth Engine Functionalities

## Variable Declaration

```JavaScript 
// Declare a variable with the keyword "var"
var variableName = ee.Thing('variableValue');

```
A container object (usually in the form `ee.Thing`) is wrapper of a native `JavaScript` object to a Googleâ€™s servers object



## Displaying a Variable or It's Metadata


```JavaScript 
// use the print() to display info in the console
print(variableName);
```
The `print` function accepts a number of inputs and displays them or the metadata associated in the console. In Earth Engine, `print` is limited to 5000 elements. Use print to debug your code


## Center the Map

```JavaScript 
Map.setCenter(lon, lat, zoomLevel);
```
Zoom level varies from 0 (no zoom) to 20 (highest zoom level)

## Add a Layer to the Map
```JavaScript 
Map.addLayer(variableName, displayParamsDict, displayName);
```

Features are geometries associated with specific properties
Feature collections are groups of features.

### Functions
A function is a set of instructions to perform a specific task:

```JavaScript 
function functionName(Arguments) {
  result = ee.Compute(Arguments);
  return result
}
```

### Function Call
```JavaScript 
var result = functionName(input);
```
This is also known as `unknown function` as opposed to `know functions` as they are assigned to a variable (compare with the first function definition)

### Mapping a Function Over a Collection
```JavaScript 
var result = ee.Collection.map(functionName);
```
Mapping a function over a collection applies the function to every element in the collection.

### Common Operations on Geometries
- Finding the area of a geometry
```JavaScript 
var geoArea = ee.geometry.area();
```

 - `.eq`  - equal
 - `.neq` - not equal
 - `.gt`  - greater than
 - `.gte` - greater than or equal to
 - `.lt`  - less than
 - `.lte` - less than or equal to

### Filter by Date Range
```JavaScript 
var dateFilter = ee.Filter.calendarRange(startDate, endDate);
```

### Filter by Particular Days of the Year
```JavaScript 
var dayFilter = ee.Filter.dayOfYear(startDay, endDay);
```

### Filtering by a Bounding Area
```JavaScript 
var boundsFilter = ee.Filter.bounds(geometryOrFeature);
```

### Combine and Inverse Filters
```JavaScript 
var newFilterAnd = ee.Filter.and(listOfFilters);
var newFilterOr = ee.Filter.or(listOfFilters);
var inverseFilter = ee.Filter.not(filter);
```



## Operations on Images
### Select an Image Band
```JavaScript 
var band = ee.image.select(bandName);
```

### Create Masks
```JavaScript 
var mask = ee.image.eq(value);
mask = mask.updateMask(mask);
``` 
The first step creates an image of `0s` and `1s` based on the condition. The second updates the mask, i.e., turns the 0s into masked values

### Apply Image Masks
```JavaScript 
var maskedImage = ee.image.updateMask(mask);
```


> or   
`Export.image.toCloudStorage()   
Export.image.toAsset()   
Export.table.toDrive()
Export.table.toCloudStorage()
Export.video.toCloudStorage()
Export.video.toDrive()`

### Bonus: Timelapse of Monthly MODIS-Aqua Rrs RGB

```JavaScript 

// Timelapse example (based on google API example);
// Create rectangle over the area of interest.
var geometry = ee.Geometry.Rectangle(136.86, 36.73, 138, 37.31);
// Add layer to map.
// Map.addLayer(geometry);
// Center to the object
Map.centerObject(geometry, 9);

// Load Aqua image collection. These are global L3 images
var aquaCollection = ee.ImageCollection('NASA/OCEANDATA/MODIS-Aqua/L3SMI')
  // Filter by date
  .filterDate('2019-01-01', '2020-01-01')
  // Select RGB Rrs
  .select('chlor_a')
  // take the log10 of each image 
  .map(function(image){
    return image.log10();
  });

// Composite VisParams
var visComp = {min: -2, max: 1.5, palette: 'blue, cyan, green, yellow, red'};
// var composite = data.mean().log10().visualize(vis);

var monthCollection = ee.ImageCollection(
  ee.List.sequence(2019, 2019).map(
    // 2003-2019
    function (year) {
      // map the time series into monthly means
      return ee.List.sequence(1, 12).map(
        function (month) {
          // Get Month Composite, Jan-Dec
          var date = ee.Date.fromYMD(year, month, 1);
          // filter month images
          return aquaCollection.filterDate(
            date, date.advance(1, 'month')
          ).mean().visualize(visComp);
        });
    }).flatten());

Map.addLayer(monthCollection.first(), {});
// var mask = monthCollection.mean().mask();
// mask = mask.updateMask(mask);

Export.video.toDrive({
  collection: monthCollection,
  // .map(
  //   function (image) {
  //     return image.updateMask(mask);
  //   }),
  // Name of file.
  description: 'ECSTimelapse',
  // Quality of video.
  dimensions: 720,
  // FPS of video.
  framesPerSecond: 8,
  // Region of export.
  region: geometry
});
// Check Tasks and run the task to get the timelapse

```
